```r
library(datasets)
install.packages("DescTools")
library(DescTools)
install.packages('tseries')
```

```r
data(iris)
summary(iris)

species<-iris$Species
Sepal_length<-iris$Sepal.Length
setosa_Sepal_length<-Sepal_length[which(species=="setosa")]
virginica_Sepal_length<-Sepal_length[which(species=="virginica")]
```

```r
t.test(setosa_Sepal_length,alternative = "two.sided", mu=5.843)
```

```r
(mean(setosa_Sepal_length)-5.843)/sqrt(var(setosa_Sepal_length)/50)
qt(0.975,49)
qt(0.25,49)
```

Sample size estimation

```r
effekt<-0.3
alpha<-0.05
beta<-0.20
n<-ceiling((qnorm(1-alpha)+qnorm(1-beta))^2/effekt^2)
n
n<-ceiling((qt(1-alpha,68)+qt(1-beta,68))^2/effekt^2)
n=71
n<-ceiling((qt(1-alpha,n-1)+qt(1-beta,n-1))^2/effekt^2)
n
```

# Paired two sample t-test

```r
t.test(setosa_Sepal_length,virginica_Sepal_length, mu=0,alternative="two.sided")
t.test(setosa_Sepal_length,virginica_Sepal_length, mu=0,alternative="greater")
```
